# Task
Performing a complete Exploratory Data Analysis (EDA) on the dataset provided named "amazon.csv".

## Load the data

### Subtask:
Load the dataset into a pandas DataFrame and display the first 5 rows to get a glimpse of the data.


**Reasoning**:
Import the pandas library, load the dataset into a DataFrame, and display the first 5 rows as requested in the instructions.



```python
import pandas as pd

df = pd.read_csv('/content/amazon.csv')
display(df.head())
```

## Initial data inspection

### Subtask:
Check the shape, column names, data types, and missing values of the dataset.


**Reasoning**:
Print the shape, column names, data types, and missing values of the dataframe to get an initial understanding of the data structure and identify any potential data quality issues.



```python
print("Shape of the DataFrame:")
print(df.shape)
print("\nColumn names and their data types:")
df.info()
print("\nNumber of missing values per column:")
print(df.isnull().sum())
```

## Summary statistics

### Subtask:
Display summary statistics for numeric columns.


**Reasoning**:
Identify the columns that should be numeric, clean them by removing non-numeric characters, convert them to a numeric type, and then calculate and display their summary statistics.



```python
import numpy as np

# Identify numeric columns that are currently objects
numeric_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

# Clean and convert numeric columns
for col in numeric_cols:
    if col in ['discounted_price', 'actual_price']:
        df[col] = df[col].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
        df[col] = pd.to_numeric(df[col])
    elif col == 'discount_percentage':
        df[col] = df[col].astype(str).str.replace('%', '', regex=False)
        df[col] = pd.to_numeric(df[col])
    elif col == 'rating':
        # Handle potential non-numeric values in rating column
        df[col] = pd.to_numeric(df[col], errors='coerce')
    elif col == 'rating_count':
        # Handle potential non-numeric values and NaNs in rating_count column
        df[col] = df[col].astype(str).str.replace(',', '', regex=False).str.replace('nan', str(np.nan), regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')


# Display summary statistics for numeric columns
display(df[numeric_cols].describe())
```

## Visualize numeric distributions

### Subtask:
Create histograms for all numeric columns to visualize their distributions.


**Reasoning**:
Create histograms for each numeric column to visualize their distributions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

## Visualize numeric boxplots

### Subtask:
Create boxplots for all numeric columns to identify potential outliers.


**Reasoning**:
Generate boxplots for each numeric column to identify potential outliers as per the instructions.



```python
numeric_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']

for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, y=col)
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()
```

## Visualize categorical counts

### Subtask:
Create countplots for all categorical/discrete columns to show their distributions.


**Reasoning**:
Identify categorical or discrete columns suitable for countplots and generate a countplot for each.



```python
categorical_cols = ['rating'] # Rating can be considered discrete/categorical as it's on a fixed scale

for col in categorical_cols:
    plt.figure(figsize=(12, 6))
    sns.countplot(data=df, x=col, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

## Correlation analysis

### Subtask:
Calculate and visualize the correlation matrix for numeric features using a heatmap.


**Reasoning**:
Calculate the correlation matrix for the selected numeric columns and visualize it using a heatmap to understand the relationships between these features.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the numeric columns
numeric_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
df_numeric = df[numeric_cols]

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

Create scatter plots of important features against the target variable "stress_level".


**Reasoning**:
Create scatter plots for the specified highly correlated numeric features to visualize their relationships.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of Actual Price vs. Discounted Price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='actual_price', y='discounted_price')
plt.title('Scatter Plot of Actual Price vs. Discounted Price')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()

# Scatter plot of Discounted Price vs. Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='discounted_price', y='rating')
plt.title('Scatter Plot of Discounted Price vs. Rating')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.show()
```

## Categorical vs. target analysis

### Subtask:
Create bar plots to visualize the relationship between categorical features and the target variable.


**Reasoning**:
Extract the main category from the 'category' column, calculate the average discounted and actual prices for each main category, and then plot these averages using bar plots.



```python
# Extract the main category
df['main_category'] = df['category'].apply(lambda x: x.split('|')[0] if isinstance(x, str) else 'Unknown')

# Analyze the number of unique main categories
print("Number of unique main categories:", df['main_category'].nunique())
print("\nTop 10 most frequent main categories:")
print(df['main_category'].value_counts().head(10))

# If the number of main categories is large, select the top N
top_n = 10
top_categories = df['main_category'].value_counts().nlargest(top_n).index

# Filter the DataFrame to include only the top categories
df_top_categories = df[df['main_category'].isin(top_categories)]

# Calculate the average discounted and actual prices for each main category
average_prices_by_category = df_top_categories.groupby('main_category')[['discounted_price', 'actual_price']].mean().reset_index()

# Sort the categories by average discounted price for better visualization
average_prices_by_category = average_prices_by_category.sort_values(by='discounted_price', ascending=False)

# Create bar plot for average discounted price
plt.figure(figsize=(12, 7))
sns.barplot(data=average_prices_by_category, x='main_category', y='discounted_price', palette='viridis')
plt.title(f'Average Discounted Price by Main Category (Top {top_n})')
plt.xlabel('Main Category')
plt.ylabel('Average Discounted Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Sort the categories by average actual price for better visualization
average_prices_by_category = average_prices_by_category.sort_values(by='actual_price', ascending=False)

# Create bar plot for average actual price
plt.figure(figsize=(12, 7))
sns.barplot(data=average_prices_by_category, x='main_category', y='actual_price', palette='viridis')
plt.title(f'Average Actual Price by Main Category (Top {top_n})')
plt.xlabel('Main Category')
plt.ylabel('Average Actual Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
